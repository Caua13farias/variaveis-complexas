### [Domain coloring](https://en.wikipedia.org/wiki/Domain_coloring)

Essa é uma técnica de plotagem interessante. Nela, o módulo e a fase são plotados como matrizes de cores diferentes na mesma figura. Em geral se usa a matriz de saturação para o módulo e a de matiz para a fase. 

Há um pacote chamado [Cplot](https://pypi.org/project/cplot/) que torna a conversão das cores e criação de gráficos complexos bastante trivial. 

```python
import numpy as np
import cplot

def f(z):
    return np.sin(z**3) / z

plt = cplot.plot(
    f,
    (-2.0, +2.0, 400),
    (-2.0, +2.0, 400),
    # abs_scaling=lambda x: x / (x + 1),  # how to scale the lightness in domain coloring
    # contours_abs=2.0,
    # contours_arg=(-np.pi / 2, 0, np.pi / 2, np.pi),
    # emphasize_abs_contour_1: bool = True,
    # add_colorbars: bool = True,
    # add_axes_labels: bool = True,
    # saturation_adjustment: float = 1.28,
    # min_contour_length = None,
    # linewidth = None,
)
plt.show()
```
